```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'ActiveUsers.csv'
data = pd.read_csv(file_path)


print(data.head())


weekly_new_users = {}
weekly_resurrected_users = {}
weekly_churned_users = {}
weekly_active_users = {}


for week in data.columns[1:]:  
    active_users = data[week].notna()
    previous_week = data.columns[data.columns.get_loc(week) - 1]
    
    if week == 'w1':
        weekly_new_users[week] = active_users.sum()
        weekly_resurrected_users[week] = 0
        weekly_churned_users[week] = 0
    else:
        previous_active_users = data[previous_week].notna()
        new_users = active_users & ~previous_active_users
        resurrected_users = active_users & previous_active_users
        churned_users = previous_active_users & ~active_users
        
        weekly_new_users[week] = new_users.sum()
        weekly_resurrected_users[week] = resurrected_users.sum()
        weekly_churned_users[week] = churned_users.sum()
    
    weekly_active_users[week] = active_users.sum()


quick_ratio = {week: (weekly_new_users[week] + weekly_resurrected_users[week]) / weekly_churned_users[week] if weekly_churned_users[week] != 0 else 0 for week in weekly_new_users.keys()}


metrics_df = pd.DataFrame({
    'New Users': weekly_new_users,
    'Resurrected Users': weekly_resurrected_users,
    'Churned Users': weekly_churned_users,
    'Active Users': weekly_active_users,
    'Quick Ratio': quick_ratio
})

plt.figure(figsize=(14, 8))
plt.plot(metrics_df.index, metrics_df['New Users'], label='New Users')
plt.plot(metrics_df.index, metrics_df['Resurrected Users'], label='Resurrected Users')
plt.plot(metrics_df.index, metrics_df['Churned Users'], label='Churned Users')
plt.plot(metrics_df.index, metrics_df['Quick Ratio'], label='Quick Ratio', linestyle='--')

plt.xlabel('Week')
plt.ylabel('Count / Quick Ratio')
plt.title('Weekly Growth Accounting')
plt.legend()
plt.grid(True)
plt.show()
```
